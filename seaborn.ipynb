```python
# Seaborn
```

```python
import seaborn as sns
df=sns.load_dataset("tips")
df
```

```python
df.corr() # to find the correlation , takes only numeric values
```

```python
sns.heatmap(df.corr())  # heatmap is the function to display the correlation
```

```python
# jointplot
sns.jointplot(x="tip",y="total_bill",data=df)
```

```python
# pairplot
sns.pairplot(df,hue='day') # plot dataframe as it is
```

```python
# distribution plot
sns.distplot(df['tip'],kde=False,bins=3)
```

```python
# countplot ,used to calculate each male and female count

sns.countplot(x='sex',data=df)
```

```python
# horizontally plot
sns.countplot(y='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df,hue='sex')
```

```python
sns.countplot(x='time',data=df,hue='smoker')
```

```python
# boxplot
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
# violinplot
sns.violinplot(x='total_bill',y='day',data=df)
```
